```python
df=pd.read_csv('test.csv')
#reading test file
```

```python
list1=data.columns
list2=df.columns
#making list of columns
```

```python
for i in range(0,data.shape[1]):
    if data[list1[i]].isna().sum()>=700:
        data=data.drop(list1[i],axis=1)
        df=df.drop(list2[i+1],axis=1)
#dropping columns which have moore then 50% of Nan values
#optional
```

```python
list1=data.columns
list2=df.columns
#defining updated column list 
```

```python
df.shape
```

```python
def function(colu,frame):
    if len(frame[colu].unique())==2:
        return (pd.get_dummies(frame[colu],drop_first=True))
    return (pd.get_dummies(frame[colu],drop_first=False))
#getting columns all numericals, removing string column data
```

```python
for i in range(0,data.shape[1]):
    if data[list1[i]].dtype=='O':
        data=pd.concat([data,function(list1[i],data)],axis=1)
        data=data.drop(list1[i],axis=1)
#concatenating new numerical columns
```

```python
data.shape
```

```python
df.head()
```

```python
for i in range(0,df.shape[1]):
    if df[list2[i]].dtype=='O':
        df=pd.concat([df,function(list2[i],df)],axis=1)
        df=df.drop(list2[i],axis=1)
#concatinating new numerical columns
```

```python
df.shape
```

```python
answer=np.zeros(df.shape[0]*2).reshape(df.shape[0],2)
#making submission 
```

```python
for i in range(df.shape[0]):
    answer[i,0]=df.iloc[i,0]
#defining ids in submission
```

```python
data.shape,df.shape
```

```python
data = data.loc[:,~data.columns.duplicated()]
df = df.loc[:,~df.columns.duplicated()]
#removing dulicate column values
```

```python
data.shape,df.shape
```

```python
list1=data.columns
list2=df.columns
#defining updated list again
```

```python
list3=[]
#new list
```

```python
for i in range(0,len(list1)):
    s=True
    for j in range(0,len(list2)):
        if list1[i]==list2[j] or list1[i]=='SalePrice':
            s=False
            break
    if s==True:
        data=data.drop(list1[i],axis=1)
#removing mutually exclusive columns
```

```python
data.shape,df.shape
```

```python
list1=data.columns
list2=df.columns
#defining updated list again
```

```python
for i in range(0,len(list1)):
    sd=0
    for j in range(0,data.shape[0]):
        if data[list1[i]][j]==0:
            sd=sd+1
    if sd>=1000:
        data=data.drop(list1[i],axis=1)
        list3.append(list1[i])
#removing columns with too much 0s to prevent bias
```

```python
for i in range(0,len(list3)):
    df=df.drop(list3[i],axis=1)
#dropping too much 0s column from even test set
```

```python
data.shape,df.shape
```

```python
list1=data.columns
list2=df.columns
list4=[]
#defining list again
```

```python
for i in range(0,len(list1)):
    if data[list1[i]].isna().sum()>=1000:
        data=data.drop(list1[i],axis=1)
        list4.append(list1[i])
#removing columns with too much nan values
```

```python
for i in range(0,len(list4)):
    df=df.drop(list4[i],axis=1)
#dropping too much nan columns
```

```python
data=data.fillna(value=0)
df=df.fillna(value=0)
#filling nan values 
```

```python
from sklearn.preprocessing import  MinMaxScaler
#to standardise the answer
```

```python
scaler=MinMaxScaler()
```

```python
from sklearn.model_selection import  train_test_split
```

```python
X=data.drop('SalePrice',axis=1)
Y=data['SalePrice']
#splitting data
```

```python
X=scaler.fit_transform(X)
#scaling features
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,train_size=0.7)
#splitting
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
#importing modules
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=50)
#defining early stop
```

```python
model=Sequential()
model.add(Dense(53,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='relu'))
model.compile(optimizer='adam',loss='mse')
#defining model
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=10000,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=early_stop
          )
          #fitting the model
```

```python
locprediction=model.predict(X_test)
#predicting on x_test
```

```python
loga=np.log(locprediction)
logb=np.log(y_test)
#taking log
```

```python
from sklearn.metrics import  mean_squared_error
```

```python
error=mean_squared_error(loga,logb)
#mean squared error
```

```python
error=np.sqrt(error)
#square root of mean squared error of their log is the criteria 
```

```python
error
```
